## CONCEITOS INTRODUTÓRIOS

### O que é Aprendizado de Máquina (Machine Learning)?

- Machine Learning é um subcampo da Inteligência Artificial. É o estudo e construção de algoritmos que podem aprender a partir de dados e fazer previsões. O aspecto iterativo do aprendizado de máquina é importante porque, conforme os modelos são expostos a novos dados, eles são capazes de se adaptar de forma independente.


- É um método de análise de dados que automatiza o desenvolvimento de modelos analíticos. Usando algoritmos que aprendem iterativamente a partir de dados, o aprendizado de máquina permite que os computadores encontrem insights através do reconhecimento de padrões.


- Inteligêncial Artificial   =>   Machine Learning   =>   Deep Learning
     
     
### Tipos de Aprendizagem de Máquina:

- <b>Aprendizagem supervisionada:</b> Previsão de valores ou classes. Os dados de treino precisam conter os valores de entrada e saída, para que o modelo aprenda como gerar a saída correta, a partir de novos dados de entrada.


- <b>Aprendizagem não supervisionada:</b> Identificação de grupos (clusters) de dados. Os dados de treino contém apenas os dados de entrada.


- <b>Aprendizagem por reforço</b>


### Aprendizagem Supervisionada:

- É o termo usado sempre que o programa é "treinado" sobre um conjunto de dados pré-definidos.

- Tipos de Aprendizagem Supervisionada:
  - Classificação
  - Regressão

### Aprendizagem Não-Supervisionada:

- Alguns sistemas de recomendação que você encontra na internet sob a forma de automação de marketing são baseados neste tipo de aprendizagem.


- O aprendizado não-supervisionado é usado com dados que não possuem rótulos históricos, ou seja, nós não temos variáveis target (as variáveis de saída) para serem estimadas.


- Por exemplo, algoritmos desta categoria, podem identificar segmentos de clientes com atributos semelhantes que podem ser tratados de modo semelhante em campanhas de marketing.


- Termo usado quando um programa pode automaticamente encontrar padrões e relações em um conjunto de dados, mesmo sem saber quais são as possíveis saídas.


### Treinamento, validação e teste:


- O processo de aprendizagem dos algoritmos de Machine Learning começa com a criação de subsets dos seus dados, são os chamados dados de treino e dados de teste.


- Em geral:
  - 75% a 70% => Dados de treino
  - 20% => Validação
  - 10% => Dados de teste
  
- É recomendável realizar a separação de forma aleatória, independente da ordenação inicial dos dados.


- Pequena regra:
  - Quando n > 10.000 pode-se criar com confiança um conjunto de dados aleatoriamente divididos.
  - Quando n < 10.000 deve-se comparar as estatísticas básicas (média, moda, mediana, variância e desvio-padrão), o que ajudará a compreender se o conjunto de dados para teste é adequado.
  
### Cross-Validation:

- A Validação Cruzada (Cross-Validation) é uma técnica para avaliar a capacidade de generalização de um modelo de Machine Learning a partir de um conjunto de dados. Esta técnica é amplamente empregada em problemas onde o motivo da modelagem é a predição. Busca-se então estimar o quão preciso é este modelo na prática, ou seja, o seu desempenho para um novo conjunto de dados.

### O que é um modelo preditivo?

- As técnicas de Machine Learning permitem a construção de algoritmos que podem aprender a fazer previsões sobre os dados.


- Em outras palavras, o modelo preditivo é uma função matemática que, aplicada a uma massa de dados, consegue identificar padrões ocultos e prever o que poderá ocorrer.


- Etapas da construção de um modelo preditivo:
  - Coleta de dados
  - Exploração e preparação
  - Treinamento do modelo
  - Avaliação do modelo
  - Otimização do modelo
  - Deploy do modelo em produção


- O trabalho do cientista de dados é buscar sempre o melhor modelo possível para suas previsões.


- O modelo pode ser implementado para resolver o problema de negócio para o qual ele foi desenvolvido.


- Se um algoritmo de Machine Learning não for capaz de generalizar uma função matemática que faça previsões sobre novos conjuntos de dados, ele não está aprendendo nada e sim memorizando os dados.

- E para generalizar a função que melhor resolve o problema, os algoritmos de Machine Learning se baseiam em 3 componentes:
  - Representação
  - Avaliação
  - Otimização
  
- Nenhum algoritimo único, ou uma combinação de algoritmos, é 100% preciso o tempo todo.

### Overfitting x Underfitting:

- <b>Overfitting</b>:
   - Um cenário de overfitting ocorre quando, nos dados de treino, o modelo tem um desempenho excelente, porém quando utilizamos os dados de teste o resultado é ruim.
   - Podemos entender que, neste caso, o modelo aprendeu tão bem as relações existentes no treino, que acabou apenas decorando o que deveria ser feito, e ao receber as informações das variáveis preditoras nos dados de teste, o modelo tenta aplicar as mesmas regras decoradas, porém com dados diferentes esta regra não tem validade, e o desempenho é afetado. É comum ouvirmos que neste cenário o modelo treinado não tem capacidade de generalização.


- <b>Underfitting</b>:
  - Neste cenário o desempenho do modelo já é ruim no próprio treinamento. O modelo não consegue encontrar relações entre as variáveis e o teste nem precisa acontecer. Este modelo já pode ser descartado, pois não terá utilidade.
  - Vamos entender estes conceitos com exemplos práticos e visuais. Na figura abaixo vemos uma distribuição de dados entre as variáveis x e y. Consideremos que estes são os dados que temos a nossa disposição para treinar e testar o modelo.
  
### Elementos do processo de aprendizagem:

- Para que ocorra a aprendizagem é preciso que:
  - Um padrão exista.
  - Não exista um único modelo matemático que explique o modelo.
  - Dados estejam disponíveis.
  
  
- Elementos do preocesso de aprendizagem:
  - Input (y): Dados dos clientes
  - Output (x): Decisão => Crédito (Sim / Não)
  - Função Alvo (f: x => y): Representação do relacionamento - Função matemática desconhecida
  - Dados: Dados históricos
  - Hipótese (g: x=> y): Função a ser descoberta pelo algoritmo.
  
  
- Espaço de Hipóteses + Algporitimo de Aprendizagem = Modelo de Aprendizagem

## TREINAR O MODELO EM TREINO E TESTE

```python
from sklearn.model_selection import train_test_split
```

```python
# Seleção das variáveis preditoras (Feature Selection)
atributos = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
```

```python
# Variável a ser prevista
atrib_prev = ['Outcome']
```

```python
# Criando objetos
X = df[atributos].values
Y = df[atrib_prev].values
```

```python
X
```

```python
Y
```

```python
# Definindo a taxa de split
split_test_size = 0.30
```

```python
# Criando dados de treino e de teste
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size = split_test_size, random_state = 42) 
```

### Tratando valores mssing - imputing

Substituir os valores iguais a zero, pela média de dados.

```python
from sklearn.impute import SimpleImputer
```

```python
# Criando objeto
preenche_0 = SimpleImputer(missing_values = 0, strategy = "mean")

# Substituindo os valores iguais a zero, pela média dos dados
X_treino = preenche_0.fit_transform(X_treino)
X_teste = preenche_0.fit_transform(X_teste)
```

```python
X_treino
```

## TREINANDO O MODELO COM NAIVE BAYS 

```python
# Utilizando um classificador Naive Bays
from sklearn.naive_bayes import GaussianNB
```

```python
# Criando o modelo preditivo
modelo_v1 = GaussianNB()
```

```python
# Treinando o modelo
modelo_v1.fit(X_treino, Y_treino.ravel())
```

### Verificando a exatidão no modelo nos dados de treino

```python
from sklearn import metrics
```

```python
nb_predict_train = modelo_v1.predict(X_treino)
```

```python
print("Exatidão (Accuracy): {0: .4f}".format(metrics.accuracy_score(Y_treino, nb_predict_train)))
print()
```

### Verificando a exatidão no modelo nos dados de treino

```python
nb_predict_test = modelo_v1.predict(X_teste)
```

```python
print("Exatidão (Accuracy): {0: .4f}".format(metrics.accuracy_score(Y_teste, nb_predict_test)))
print()
```

### Métricas

```python
# Criando uma Confusion Matrix
print("Confusion Matrix")

print("{0}".format(metrics.confusion_matrix(Y_teste, nb_predict_test, labels = [1, 0])))
print("")

print("Classification Report")
print(metrics.classification_report(Y_teste, nb_predict_test, labels = [1, 0]))
```

## OTIMIZAÇÃO DO MODELO COM RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo_v2 = RandomForestClassifier(random_state = 42)
modelo_v2.fit(X_treino, Y_treino.ravel())
```

```python
# Verificar nos dados de treino
rf_predict_train = modelo_v2.predict(X_treino)
print("Exatidão (Accuracy): {0: .4f}".format(metrics.accuracy_score(Y_treino, rf_predict_train)))
```

```python
# Verificar nos dados de teste
rf_predict_test = modelo_v2.predict(X_teste)
print("Exatidão (Accuracy): {0: .4f}".format(metrics.accuracy_score(Y_teste, rf_predict_test)))
```

```python
# Criando uma Confusion Matrix
print("Confusion Matrix")

print("{0}".format(metrics.confusion_matrix(Y_teste, rf_predict_test, labels = [1, 0])))
print("")

print("Classification Report")
print(metrics.classification_report(Y_teste, rf_predict_test, labels = [1, 0]))
```
